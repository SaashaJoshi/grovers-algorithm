```python
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
qc = QuantumCircuit(7, 2)

# Compare 2-qubit strings 

# String 1 (01)
qc.x(1)

# String 2 (10)
qc.x(2)
# qc.x(2)    # (11)
```

```python
qc.draw('mpl')
```

```python
# compare_lsb 
control_lsb = [1, 3]
# compare_msb 
control_msb = [0, 2]

qc.mct(control_msb, 4)
qc.mct(control_lsb, 5)
```

```python
qc.draw('mpl')
```

```python
qc.measure([4, 5], [1, 0])
qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
results = execute(qc, backend, shots = 1024).result()
counts = results.get_counts()
print(counts)
plot_histogram(counts)

# Both bits different: {00}
# MSB same: {10}
# LSB same: {01}
# Both same: {11}
```
